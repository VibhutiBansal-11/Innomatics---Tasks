Hypothesis Testing

Step 1 $\implies$ $$ H_0(Status Quo) \implies \mu \leq 2.5 \\ H_1(Bold Claim) \implies \mu > 2.5 $$

Step 2 $\implies$ $$ Collecting \ samples \ from \ data $$

Step 3 $\implies$ $$ Calculating \ \mu_x \ and \ S^2 \ for \ the \ sample $$

Step 4 $\implies$ $$ Setting \ confidence \ level \ to \ 95 \% \\ \alpha = 0.05 $$

Step 5 $\implies$ $$ Computing \ t \ score \ and \ testing \ test \ staticstics \ with t \ and \ p \ test  $$

```python
samplesSalary = dataForComputerScienceAndDesignation.loc[:, "Salary"]
samplesSalary.describe()
```

```python
population_mean = 250000
samples_size = len(samplesSalary)
samples_mean = samplesSalary.mean()
samples_std = samplesSalary.std()
```

```python
alpha = 1 - 0.95
print(alpha)
```

```python
def t_score(population_mean, sample_mean, samples_size, samples_std):
  numerator = population_mean - samples_mean
  denominator = samples_std / (samples_size ** 0.5)
  return numerator / denominator
```

```python
t_calculated = t_score(population_mean, samples_mean, samples_size, samples_std)
print(t_calculated)
```

```python
from scipy.stats import t
```

```python
t_critical = t.ppf(1-alpha, df = (len(dataset.columns) - 1))
t_critical
```

```python
if t_calculated > t_critical :
  print("Reject Null Hypothesis")
else:
  print("Failed to reject null hypothesis")
```

Chi Square Testing

Step 1 $\implies$ $$ H_0(Status \ Quo) \implies Gender \ and \ Specialization \ are \ not \ related \\ H_1(Bold \ Claim) \implies Gender \ and \ Specialization \ are \ related $$

Step 2 $\implies$ $$ Collecting \ samples \ from \ data $$

Step 3 $\implies$ $$ Calculating \ chi2 \ score \ for \ the \ sample $$

Step 4 $\implies$ $$ Setting \ confidence \ level \ to \ 95 \% \\ \alpha = 0.05 $$

Step 5 $\implies$ $$ Testing \ test \ staticstics \ with \ chi2 \ test  $$

```python
dataset["Gender"].value_counts()
```

```python
pd.crosstab(dataset["Specialization"], dataset["Gender"], margins=True)
```

```python
observed = pd.crosstab(dataset["Specialization"], dataset["Gender"])
observed
```

```python
from scipy.stats import chi2, chi2_contingency
```

```python
chi2_contingency(observed)
```

```python
chi2_test_stat = chi2_contingency(observed)[0]
pval = chi2_contingency(observed)[1]
df = chi2_contingency(observed)[2]
```

```python
confidence_level = 0.95
alpha = 1 - confidence_level
chi2_critical = chi2.ppf(1-alpha, df)
chi2_critical
```

```python
if chi2_test_stat < chi2_critical:
  print("Reject Null Hypothesis")
else:
  print("Failed to reject null hypothesis")
```
