## Mean
Mean is the sum of a collection of numbers divided by the count of numbers in the collection.

$$ Mean = \frac{1}{n} \sum_{i=1}^{n} A_{i}$$

```python
mean = {}
for i in numerical_col:
    total = 0
    for j in data.iloc[:,i]:
        total+=j
    mean[i]=total/data.shape[0]
        
    print(data.columns[i],":",total/data.shape[0])
    
```

## Median
Median is the value separating the higher half from the lower half of a data sample, a population, or a probability distribution. For a data set, it may be thought of as "the middle" value.
$$
Median = \left\{
    \begin{array}\\
        X[\frac{N+1}{2}] & \mbox{if N is odd }\\
        \frac{X[\frac{N}{2}]+X[\frac{N}{2}+1]}{2} & \mbox{if N is even}  \\
    \end{array}
\right.
$$

```python

for i in numerical_col:
    sort_col = data.iloc[:,i].sort_values()
    N = len(sort_col)

    if(N%2!=0):
        print(data.columns[i],":",sort_col.iloc[(N+1)//2])
    else:
     
        print(data.columns[i],":",(sort_col.iloc[N//2-1]+sort_col.iloc[N//2])/2)
```

```python
data.median()
```

## Variance
Variance is the expectation of the squared deviation of a random variable from its mean. In other words, it measures how far a set of numbers is spread out from their average value.
$$ 
\sigma ^ 2 = S^2 = \frac {\sum{(x  - \mu)^2}}{n-1}\\
\\
$$
$\sigma^2  = S^2$ = Sample Variance <br>
$x = $ Value of observation  <br>
$\mu$ =  Mean <br>
$n =$ No. of observations <br>


```python
for i in numerical_col:
    total = 0
    for j in data.iloc[:,i]:
        total+= ((j-mean[i])**2)
    print(data.columns[i],":",total/(data.shape[0]-1))
```

```python
data.var()
```

## Standard Deviation
The standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range. 
$$
\sigma = \sqrt{\frac{\sum(x - \mu)^2}{N}}
$$
$\sigma$ = 	population standard deviation<br>
$x = $ Value of observation  <br>
$\mu$ =  Mean <br>
$N =$ No. of observations <br>

```python
for i in numerical_col:
    total = 0
    for j in data.iloc[:,i]:
        total+= ((j-mean[i])**2)
    print(data.columns[i],":",(total/(data.shape[0]-1))**0.5)
```

```python
data.std()
```

## Correlation
It always lies between 0 and 1 . It helps us understand how one variable changes w.r.t other variables. <br><br>
$$
r = \frac {\sum(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sqrt{\sum(x_{i}-\bar{x})^2\sum(y_{i}-\bar{y})^2}}
$$
$r$	=	Pearson correlation coefficient <br>
$x_{i}$	=	values of the x-variable in a sample <br>
$\bar{x}$	=	mean of the values of the x-variable <br>
$y_{i}$	=	values of the y-variable in a sample<br>
$\bar{y}$	=	mean of the values of the y-variable<br>

## Normal Distribution
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-1}{2}(\frac{x - \mu}{\sigma})^2}
$$
$f(x)$	=	probability density function  <br>
$\sigma$	=	standard deviation <br>
$\mu$	=	mean <br>
[![Alt Text](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/1200px-Normal_Distribution_PDF.svg.png)](https://medium.com/@ahmedazizkhelifi)


```python
from scipy import stats
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
for i in data.columns:

    sns.displot(data=data, x=i, kde=True)
   
```

## Features of Normal Distribution



1. It follows the 68% - 95% - 99.7% rule.
In a normal distribution: 68%, 95%, and 99.7% of the values lie within one, two, and three standard deviations of the mean.

2. It has a bell-shaped curve.
3. Commonly found distribution in real-life data.
4. The mean-mode-median is in the center.
5. Higher the $\sigma^2$ flatter the curve.
6. Peak lies at $\mu$

```python
import numpy as np
x = np.random.normal(25000, 5000, size=10000)
sns.distplot(x)
```

```python
mu = x.mean()
sigma = x.std()
print(mu, sigma)
one_std_right = mu + (1 * sigma)
one_std_left = mu - (1 * sigma)
two_std_right = mu + (2 * sigma)
two_std_left = mu - (2 * sigma)
three_std_right = mu + (3 * sigma)
three_std_left = mu - (3 * sigma)
```

```python

plt.figure(figsize=(20,10))
sns.set_style()
sns.distplot(x,color='black')

plt.axvline(mu, color='black', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvspan(one_std_right,one_std_left, alpha=0.5, color='yellow',ymin=0)
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')

plt.axvline(two_std_right, color='red', label='Mean + 2SD')
plt.axvspan(one_std_right,two_std_right, alpha=0.5, color='red',ymin=0)
plt.axvspan(one_std_left,two_std_left, alpha=0.5, color='red',ymin=0)
plt.axvline(two_std_left, color='red', label='Mean - 2SD')

plt.axvline(three_std_right, color='green', label='Mean + 3SD')
plt.axvspan(two_std_right,three_std_right, alpha=0.5, color='green',ymin=0)
plt.axvspan(two_std_left,three_std_left, alpha=0.5, color='green',ymin=0)
plt.axvline(three_std_left, color='green', label='Mean - 3SD')
plt.legend()
```

## Positively Skewed Normal Distribution & Negatively Skewed Normal Distribution

$$
Skewness Level =
\left\{
    \begin{array}\\
  Highly Skewed & \mbox{if |s| > 1 } \\
        moderately skewed & \mbox{if 0.5 < |s| < 1} \\
         approximately symmetric & \mbox{|s| < 0.5}
    \end{array}
\right.
$$
<br>
<br>
 If one tail is longer than another, the distribution is skewed. These distributions are sometimes called asymmetric or asymmetrical distributions as they dont show any kind of symmetry. A left-skewed distribution has a long left tail. Left-skewed distributions are also called negatively-skewed distributions. Thats because there is a long tail in the negative direction on the number line. The mean is also to the left of the peak.A right-skewed distribution has a long right tail. Right-skewed distributions are also called positive-skew distributions. Thats because there is a long tail in the positive direction on the number line. The mean is also to the right of the peak.
 




```python
data.skew()
```

## Effect on Mean, Median and Mode due to Skewness
Generally if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean
As mean is most heavily influenced by any outliers or skewness. For distributions that have outliers or are skewed, the median is often the preferred measure of central tendency because the median is more resistant to outliers than the mean.

## QQ Plot
QQ plot is a probability plot, which is a graphical method for comparing two probability distributions by plotting their quantiles against each other. It used to check the normal nature or normality of a distribution.

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
stats.probplot(data['Mthly_HH_Income'], dist="norm", plot=plt)

plt.subplot(2, 3, 2)
stats.probplot(data['Mthly_HH_Expense'], dist="norm", plot=plt)

plt.subplot(2, 3, 3)
stats.probplot(data['No_of_Fly_Members'], dist="norm", plot=plt)

plt.subplot(2, 3, 4)
stats.probplot(data['Emi_or_Rent_Amt'], dist="norm", plot=plt)

plt.subplot(2, 3, 5)
stats.probplot(data['Annual_HH_Income'], dist="norm", plot=plt)

plt.subplot(2, 3, 6)
stats.probplot(data['No_of_Earning_Members'], dist="norm", plot=plt)

plt.show()
```

## Box Cox
A Box Cox transformation is a transformation of a non-normal dependent variables into a normal shape. 
It changes our data so that it resembles normal distribution.

```python
plt.figure(figsize=(12, 10))
plt.subplot(2, 3, 1)
stats.probplot(stats.boxcox(data['Mthly_HH_Income'])[0], dist="norm", plot=plt)
print(stats.boxcox(data['Mthly_HH_Income'])[1])
plt.subplot(2, 3, 2)
stats.probplot(stats.boxcox(data['Mthly_HH_Expense'])[0], dist="norm", plot=plt)
print(stats.boxcox(data['Mthly_HH_Expense'])[1])
plt.subplot(2, 3, 3)
stats.probplot(stats.boxcox(data['No_of_Fly_Members'])[0], dist="norm", plot=plt)
print(stats.boxcox(data['No_of_Fly_Members'])[1])
plt.subplot(2, 3,4)
stats.probplot(stats.boxcox(data['Annual_HH_Income'])[0], dist="norm", plot=plt)
print(stats.boxcox(data['Annual_HH_Income'])[1])
plt.subplot(2, 3, 5)
stats.probplot(stats.boxcox(data['No_of_Earning_Members'])[0], dist="norm", plot=plt)
print(stats.boxcox(data['No_of_Earning_Members'])[1])
plt.show()
```
